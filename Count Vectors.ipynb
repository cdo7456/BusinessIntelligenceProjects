## Sentiment Analysis

```python
df = pd.read_csv(PATH, encoding='unicode_escape')
df
```

```python
df = df[['text', 'sentiment']].copy()
df.head()
```

```python
df.sentiment.value_counts()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# define the vectorizer
vect = CountVectorizer()
# fit the vectorizer
vect.fit(df['text'])
```

```python
len(vect.vocabulary_)
```

```python
vect.vocabulary_
```

```python
X = vect.transform(df['text'])
```

```python
X
```

```python
# divide the data into training and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, df['sentiment'], test_size=.3, random_state=42)
```

```python
# build the model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=300)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred[:10]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# let's try PCA to reduce the dimension
from sklearn.decomposition import PCA, TruncatedSVD

pca = TruncatedSVD(n_components=100)
pca.fit(X_train)
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```

```python
model = LogisticRegression(max_iter=300)
model.fit(pca.transform(X_train), y_train)
y_pred = model.predict(pca.transform(X_test))

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

### Stopwords

```python
vect = CountVectorizer(stop_words = 'english')
vect.fit(df['text'])
len(vect.vocabulary_)

X = vect.transform(df['text'])

# divide the data into training and test
X_train, X_test, y_train, y_test = train_test_split(X, df['sentiment'], test_size=.3, random_state=42)

model = LogisticRegression(max_iter=300)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy_score(y_test, y_pred)
```
