```python
plt.figure(figsize=(13,4))

plt.subplot(1,2,1)
sns.histplot(x="Admitted", data=df, color='#288BA8')

plt.subplot(1, 2, 2)
sns.countplot(x= data["Enrolled"], saturation=2)
```

From the two plots above, we can see there is a difference in the number of students admitted, and the number of students enrolled.

This is not surprising as not all students who are admitted into college will end up attending. This could be cost of tuition, or a change of plans.

```python
sns.countplot(x=data['Gender'], hue=data['Admitted'], saturation=1)
plt.title("Students Admitted in College by Gender")
```

From the plot below, we can see that on average, more females are admitted into college than males.

What is surprising however, is that more more females are also not admitted into college than males. This could be due to another variable, such as the fact that more females applied to college, than more males did.

```python
# Group applicants based on SAT/ACT scores
gpa_bins = pd.cut(df['SAT/ACT'], [0, 1100, 1300])
gpa_bins
```

```python
gpa_bins = pd.cut(df['SAT/ACT'], [0, 1100, 1300])
df.pivot_table('SAT/ACT', index=gpa_bins, columns='Admitted', aggfunc='count')
```

We can see from the Pivot Table above, applicants with a SAT/ACT score from 0 to 1100 we admitted less than students with an SAT/ACT score higher than 1100.

This is not surprising, as we would expect students with higher SAT/ACT scores to be more successful with their college applications, than students with a lower SAT/ACT score.

```python
# Pivot table that shows the average SAT/ACT scores of an applicant by enrollment statu and gender
gpa_bins = pd.cut(df['SAT/ACT'], [0, 1100, 1300])
df.pivot_table('Applicant', index='Enrolled', columns=['Gender', gpa_bins], aggfunc='mean')
```

We can see from the data, that males have a have a higher average enrollment rate than females - in both GPA bins.
