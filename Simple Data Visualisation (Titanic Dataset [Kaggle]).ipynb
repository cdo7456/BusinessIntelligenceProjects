```python
# Fill in the missing values for Age
df

# First, check how many missing values are in the Age column:
df.isnull().sum()
# 177 missing values in the Age column
```

```python
df.Age.isnull().sum()
```

```python
# Fill in the missing values of Age by the median of 'Pclass' and 'Sibsp', create a new subset:
age_by_Pclass_SibSp = df.groupby(['Pclass', 'SibSp']).median()['Age']

age_by_Pclass_SibSp
```

```python
# We will manually need to fill in the values for Pclass 3, SibSp 8:
age_by_Pclass_SibSp[1].index.tolist()
age_by_Pclass_SibSp[3][8] = age_by_Pclass_SibSp[3][5]

# Read in the edited dataset
age_by_Pclass_SibSp
```

```python
# Fill in the missing Ages by using the median of Pclass and SibSp:
df['Age'].fillna(df['Age'].median(),inplace = True)
df
```

```python
# Verify that there are no longer missing values for the 'Age' column
df.isnull().sum()
```

```python
# Fill in the missing values for Cabin
df

# First check how many missing values are present for Cabin:
df.isnull().sum()
# 687 missing values for the Cabin column
```

```python
# Group the class a passenger is in with a possible cabin deck by defining it:
def cabin_set(data_cabin):
  cabin_type = []

  for i in range(len(data_cabin)):

    if data_cabin.isnull()[i] == True:
      cabin_type.append('Missing')

    else:
        cabin = data_cabin[i]
        cabin_type.append(cabin[:1])

  return cabin_type
```

```python
# Define the column
df['Cabin'] = df['Cabin'].fillna('M').astype(str).apply(lambda cabin: cabin[0])
```

```python
# Verify the values have been filled in:
df.Cabin.value_counts()
```

```python
# Fill in the missing values for Embarked
df

# First, check how many missing values for embarked
df.Embarked.isnull().sum
```

```python
# By research, I found the probability of boarding the Titanic from Southahmpton (S), was higher, so I filled the missing values with Southampton (S)
df.Embarked.fillna(df.Embarked.mode()[0], inplace = True)
df.Embarked
```

```python
# Verify values have been filled in
df.Embarked.value_counts()
```
